```python
df3 = df[df['State'].isin(states)]
avg = []
for i in df3.itertuples():
    c =[]
    c.append(i.State)
    coal = (i.CoalC2010 + i.CoalC2011 + i.CoalC2012 + i.CoalC2013 + i.CoalC2014)/5
    nat = (i.NatGasC2010 + i.NatGasC2011 + i.NatGasC2012 + i.NatGasC2013 + i.NatGasC2014)/5
    lpg = (i.LPGC2010 + i.LPGC2011 + i.LPGC2012 + i.LPGC2013 + i.LPGC2014)/5
    c.append(coal)
    c.append(nat)
    c.append(lpg)
    avg.append(c)
df6 = pd.DataFrame(avg, columns=['State', 'Coal', 'NatGas', 'LPG'])
df6.plot.bar(x='State', y=['Coal', 'NatGas', 'LPG'], title="Average Coal, NatGas and LPG consumption of States", figsize=(20,10))
plt.ylabel('BTU')
plt.show()
```

```python
df3 = df[df['State'].isin(states)]
avg = []
for i in df3.itertuples():
    c =[]
    c.append(i.State)
    Hydro = (i.HydroC2010 + i.HydroC2011 + i.HydroC2012 + i.HydroC2013 + i.HydroC2014)/5
    Geo = (i.GeoC2010 + i.GeoC2011 + i.GeoC2012 + i.GeoC2013 + i.GeoC2014)/5
    c.append(Hydro)
    c.append(Geo)
    avg.append(c)
df6 = pd.DataFrame(avg, columns=['State', 'Hydro', 'Geo'])
df6.plot.bar(x='State', y=['Hydro', 'Geo'], title="Average Hydro and Geo consumption of States", figsize=(20,10))
plt.ylabel('BTU')
plt.show()
```
